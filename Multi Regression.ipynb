```python
data=pd.read_csv("Housing clean.csv")
data.head()
```

```python
x=data[['area','parking','mainroad','stories','prefarea','bedrooms','airconditioning']]
y=data[['price']]
```

```python
encoder=LabelEncoder()
# x['furnishingstatus']=encoder.fit_transform(x.furnishingstatus)
x['mainroad']=encoder.fit_transform(x.mainroad)
x['prefarea']=encoder.fit_transform(x.prefarea)
x['airconditioning']=encoder.fit_transform(x.airconditioning)
```

```python
x
```

```python
x.corr()
```

```python
lr=LinearRegression()
```

```python
lr.fit(x,y)
```

```python
pred=lr.predict(x)
```

```python
print(r2_score(y,pred))
```
