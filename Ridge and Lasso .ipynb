```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from scipy.stats import iqr


from sklearn.preprocessing import LabelEncoder
# --to convert string into binary LabelEncoder



from sklearn.metrics import r2_score
# for gettin the score of accuracy of prediction
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge,Lasso
```

```python
data=pd.read_csv("Housing clean.csv")
data.head()
```

```python
x=data[['area','parking','mainroad','stories','prefarea','bedrooms','airconditioning']]
y=data[['price']]
```

```python
encoder=LabelEncoder()
# x['furnishingstatus']=encoder.fit_transform(x.furnishingstatus)
x['mainroad']=encoder.fit_transform(x.mainroad)
x['prefarea']=encoder.fit_transform(x.prefarea)
x['airconditioning']=encoder.fit_transform(x.airconditioning)
```

```python
# Assuming 'X' is your feature matrix and 'y' is your target variable

x_train,x_test,y_train, y_test = train_test_split(x,y,test_size=0.1,random_state=1)

```

```python
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
x_test_prediction=lr.predict(x_test)
print(r2_score(y_test,x_test_prediction))
```

```python
x_train_pred=lr.predict(x_train)
r2_score(y_train,x_train_pred)
```

```python
# Ridge
rr=Ridge(alpha=0.2)
rr.fit(x_train,y_train)
```

```python
rr_pred=rr.predict(x_test)
```

```python
r2_score(y_test,rr_pred)
```

```python
# Lasso

laso_r=Lasso(alpha=0.2)
laso_r.fit(x_train,y_train)
```

```python
x_test_pred_lasso=laso_r.predict(x_test)
r2_score(y_test,x_test_pred_lasso)
```
